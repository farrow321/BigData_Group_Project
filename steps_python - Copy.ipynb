```python
from operator import index

import pandas as pd
import time 
```

```python

start_time_overall = time.time()
elapsed_times_list = []
```

#1 loading csv file 

```python
start_time1 = time.time()

df = pd.read_csv("beer_reviews.csv")

end_time1 = time.time()
elapsed_time1 = end_time1 - start_time1
elapsed_times_list.append(elapsed_time1)
print(f"Elapsed time: {elapsed_time1} seconds")
```

#2 which beer style is rated highest review overall

```python
start_time2 = time.time()

highest_review = df[df['review_overall'] == 5]

end_time2 = time.time()
elapsed_time2 = end_time2 - start_time2
elapsed_times_list.append(elapsed_time2)
print(f"Elapsed time: {elapsed_time2} seconds")

highest_review
```

#3 Best beer per brewery

```python
start_time3 = time.time()

best_beer_per_brewery = df.sort_values(by=["review_overall", 'index'],ascending=False).groupby("brewery_name").head(1)

end_time3 = time.time()
elapsed_time3 = end_time3 - start_time3
elapsed_times_list.append(elapsed_time3)
print(f"Elapsed time: {elapsed_time3} seconds")

best_beer_per_brewery[best_beer_per_brewery['brewery_name'] == 'Pacific Coast Brewing Company']
```

#4 Average review_overall Score per Brewery and Identify the Top 5 Breweries

```python
start_time4 = time.time()

average_reviews_per_brewery = df.groupby('brewery_name')['review_overall'].mean().reset_index()
average_reviews_per_brewery.columns = ['brewery_name', 'avg_review_overall']

end_time4 = time.time()
elapsed_time4 = end_time4 - start_time4
elapsed_times_list.append(elapsed_time4)
print(f"Elapsed time: {elapsed_time4} seconds")

average_reviews_per_brewery
```

#5 most reviewed beer styles

```python
start_time5 = time.time()

most_reviewed_styles = df.groupby('beer_style').size().reset_index(name='review_count')
most_reviewed_styles = most_reviewed_styles.sort_values(by='review_count', ascending=False)

end_time5 = time.time()
elapsed_time5 = end_time5 - start_time5
elapsed_times_list.append(elapsed_time5)
print(f"Elapsed time: {elapsed_time5} seconds")

most_reviewed_styles
```

#6 Identify the Most Polarizing Beers Based on review_aroma

```python
start_time6 = time.time()

variance_aroma_by_beer = df.groupby('beer_style')['review_aroma'].var().reset_index()
variance_aroma_by_beer.columns = ['beer_style', 'review_aroma']

end_time6 = time.time()
elapsed_time6 = end_time6 - start_time6
elapsed_times_list.append(elapsed_time6)
print(f"Elapsed time: {elapsed_time6} seconds")

variance_aroma_by_beer
```

```python
end_time_overall = time.time()
elapsed_time_visualisation = end_time_overall - start_time_overall
elapsed_time_no_visualisation = sum(elapsed_times_list)
print(f"Elapsed time: {elapsed_time_no_visualisation} seconds without visualising data")
print(f"Elapsed time: {elapsed_time_visualisation} seconds with visualising data")
```
